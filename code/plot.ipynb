```python
sample='MERV - XMEV - GFGV6524AG - 24hs'
data_base_merged=data_base_merged[data_base_merged["id_simbolo"]==sample]
trade_log=trade_log[trade_log["id_simbolo"]==sample]
```

```python
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio
import pandas as pd

# --- Aseguramos fechas ---
data_base_merged['biof_fecha'] = pd.to_datetime(data_base_merged['biof_fecha'])
trade_log['biof_fecha'] = pd.to_datetime(trade_log['biof_fecha'])

# --- PNL acumulado ---
trade_log['pnl_acum'] = trade_log['pnl_total'].cumsum()

# --- Crear figura con 2 filas ---
fig = make_subplots(
    rows=2, cols=1,
    shared_xaxes=True,
    vertical_spacing=0.15,
    row_heights=[0.7, 0.3],
    subplot_titles=("Order Book L1 + Trades", "PNL Acumulado"),
    specs=[[{"secondary_y": True}], [{}]]  # <-- eje secundario en la primera fila
)

# ======================
#  Subplot 1 (Order Book con eje secundario)
# ======================

# Opción Bid L1 (eje primario)
fig.add_trace(go.Scatter(
    x=data_base_merged['biof_fecha'],
    y=data_base_merged['bi_1_precio'],
    mode='lines',
    name='Opción Bid L1',
    line=dict(color='blue', dash='dash')
), row=1, col=1, secondary_y=False)

# Opción Ask L1 (eje primario)
fig.add_trace(go.Scatter(
    x=data_base_merged['biof_fecha'],
    y=data_base_merged['of_1_precio'],
    mode='lines',
    name='Opción Ask L1',
    line=dict(color='orange', dash='dash')
), row=1, col=1, secondary_y=False)

# Underlying Bid L1 (eje secundario)
fig.add_trace(go.Scatter(
    x=data_base_merged['biof_fecha'],
    y=data_base_merged['bi_1_precio_under'],
    mode='lines',
    name='Underlying Bid L1',
    line=dict(color='green', dash='dash')
), row=1, col=1, secondary_y=True)

# Underlying Ask L1 (eje secundario)
fig.add_trace(go.Scatter(
    x=data_base_merged['biof_fecha'],
    y=data_base_merged['of_1_precio_under'],
    mode='lines',
    name='Underlying Ask L1',
    line=dict(color='red', dash='dash')
), row=1, col=1, secondary_y=True)

# Trades (sobre el eje primario de la opción)
fig.add_trace(go.Scatter(
    x=trade_log['biof_fecha'],
    y=trade_log['avg_p_opt_share'],
    mode='markers',
    name='Trades',
    marker=dict(color='black', size=8, symbol='circle')
), row=1, col=1, secondary_y=False)

# ======================
#  Subplot 2 (PNL Acumulado)
# ======================

fig.add_trace(go.Scatter(
    x=trade_log['biof_fecha'],
    y=trade_log['pnl_acum'],
    mode='lines+markers',
    name='PNL Acumulado',
    line=dict(color='purple')
), row=2, col=1)

# ======================
# Layout general
# ======================
fig.update_layout(
    title=f"{sample} Order Book + Trades + PNL Acumulado",
    xaxis=dict(
        tickformat="%Y-%m-%d %H:%M:%S.%3f"  # eje X con milisegundos
    ),
    yaxis=dict(
        title="Precio Opción",
        titlefont=dict(color="black"),
        tickfont=dict(color="black")
    ),
    yaxis2=dict(
        title="Precio Subyacente",
        titlefont=dict(color="black"),
        tickfont=dict(color="black")
    ),
    legend=dict(
        orientation="h",
        yanchor="bottom",
        y=-0.25,
        xanchor="center",
        x=0.5
    ),
    template="plotly_white",
    hovermode="x unified",
    height=800
)

# Guardar como HTML
fig.write_html("orderbook_pnl_acum.html", auto_open=True)
```
